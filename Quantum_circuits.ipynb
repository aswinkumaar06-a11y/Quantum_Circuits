```python
import cirq

# Let's define a single quantum bit (qubit) and give it a name
my_first_qubit = cirq.NamedQubit("Q1")

print("My qubit is:", my_first_qubit)
```

```python
# Create the operation (applying the X gate to our qubit)
my_operation = cirq.X(my_first_qubit)

# Place that operation inside a new circuit
my_circuit = cirq.Circuit(my_operation)

# Let's see what it looks like
print(my_circuit)
```

```python
# 1. Create the circuit with the X gate and a Measurement
circuit = cirq.Circuit(
    cirq.X(my_first_qubit),
    cirq.measure(my_first_qubit, key='result')
)

# 2. Create the simulator
simulator = cirq.Simulator()

# 3. Run the circuit 10 times
samples = simulator.run(circuit, repetitions=10)
print(circuit)
print(samples)
```

```python

# Create a circuit with a Hadamard gate
superposition_circuit = cirq.Circuit(
    cirq.H(my_first_qubit),
    cirq.measure(my_first_qubit, key='result')
)

# Run it 10 times
samples = simulator.run(superposition_circuit, repetitions=10)
print(samples)
print(superposition_circuit)
```

```python
# Create two qubits
q0 = cirq.NamedQubit("control")
q1 = cirq.NamedQubit("target")

# Build the entanglement circuit
bell_circuit = cirq.Circuit(
    cirq.H(q0),              # Put q0 into superposition
    cirq.CNOT(q0, q1),       # Link q0 and q1
    cirq.measure(q0, q1, key='result') # Measure both
)

print("--- Bell Circuit ---")
print(bell_circuit)

# Run it
samples = simulator.run(bell_circuit, repetitions=10)
print("\n--- Results ---")
print(samples)
```

```python
import cirq

# 1. Setup the Qubits
# We'll use GridQubits this time, which represent physical locations on a chip
q0 = cirq.GridQubit(0, 0)
q1 = cirq.GridQubit(0, 1)

# 2. Define the Circuit
# We create a Bell State: a 50/50 chance of both being 0 OR both being 1
bell_circuit = cirq.Circuit(
    cirq.H(q0),                    # Step 1: Put q0 into superposition (0 and 1)
    cirq.CNOT(q0, q1),             # Step 2: Entangle q1 with q0
    cirq.measure(q0, q1, key='m')  # Step 3: Measure both qubits
)

# 3. Simulate the Circuit
simulator = cirq.Simulator()
result = simulator.run(bell_circuit, repetitions=10)

# 4. Display the results
print("Circuit Diagram:")
print(bell_circuit)
print("\nMeasurement Results:")
print(result)
```

```python
import cirq
import numpy as np

# Create a qubit and a circuit
q = cirq.GridQubit(0, 0)

circuit = cirq.Circuit(cirq.H(q)) # Hadamard puts it on the X-axis

# Use the simulator to get the "State Vector" (the arrow's coordinates)
simulator = cirq.Simulator()
result = simulator.simulate(circuit)

# Print the state vector
print("State Vector:")
print(result.final_state_vector)

# Cirq can also show us the (x, y, z) coordinates on the sphere
bloch_vector = cirq.bloch_vector_from_state_vector(result.final_state_vector, index=0)
print("\nBloch Sphere Coordinates (x, y, z):")
print(np.around(bloch_vector, 4))
```

```python
import cirq
import numpy as np

q = cirq.GridQubit(0, 0)

# Apply an X gate raised to the power of 0.25 (45-degree rotation)
circuit = cirq.Circuit(cirq.X(q)**0.25)

simulator = cirq.Simulator()
result = simulator.simulate(circuit)

# Get the Bloch vector (x, y, z)
bloch_vector = cirq.bloch_vector_from_state_vector(result.final_state_vector, index=0)

print("Circuit:")
print(circuit)
print("\nBloch Sphere Coordinates (x, y, z):")
print(np.around(bloch_vector, 4))
```
