```python
import cirq

# Let's define a single quantum bit (qubit) and give it a name
my_first_qubit = cirq.NamedQubit("Q1")

print("My qubit is:", my_first_qubit)
```

```python
# Create the operation (applying the X gate to our qubit)
my_operation = cirq.X(my_first_qubit)

# Place that operation inside a new circuit
my_circuit = cirq.Circuit(my_operation)

# Let's see what it looks like
print(my_circuit)
```

```python
# 1. Create the circuit with the X gate and a Measurement
circuit = cirq.Circuit(
    cirq.X(my_first_qubit),
    cirq.measure(my_first_qubit, key='result')
)

# 2. Create the simulator
simulator = cirq.Simulator()

# 3. Run the circuit 10 times
samples = simulator.run(circuit, repetitions=10)

print(samples)
```

```python
# Create a circuit with a Hadamard gate
superposition_circuit = cirq.Circuit(
    cirq.H(my_first_qubit),
    cirq.measure(my_first_qubit, key='result')
)

# Run it 10 times
samples = simulator.run(superposition_circuit, repetitions=10)
print(samples)
```
